# QM9 compounds

```python
qm9 = pd.read_table('../graph-neural-networks/mpnn-data/qm9/dsgdb9nsd.inchi',header=None,names=['cpd','InChI'])
qm9['cpd'] = qm9['cpd'].str.split(':').str.get(0)
```

# Friendly MetaCyc Uniprot proteins sequences

```python
import pandas as pd
import os
os.listdir('Dataset')
```

```python
rxns_of_enz =pd.read_table('Dataset/Enzymes_of_a_reaction_from_Friendly_Metacyc_reactions.tab',index_col=0)
rxns_of_enz['rxns'] = rxns_of_enz['Matches'].str.split(' // ')
rxns_of_enz
```

```python
friendly_rxns = pd.read_table('Dataset/friendly-metacyc-rxns.tab',index_col='FrameId')
friendly_rxns
```

```python
friendly_enzymes = []
for enz in rxns_of_enz.index:
    if len(set(rxns_of_enz.loc[enz,'rxns']) & set(friendly_rxns.index)) > 0:
        friendly_enzymes.append( enz )
rxns_of_enz.loc[friendly_enzymes]
```

```python
m2u = pd.read_table('Dataset/UniProt_from_All_instances_of_Polypeptides_in_MetaCyc.tab',index_col='Matches')
uniprot_of_friendly_enz = rxns_of_enz.join(m2u,how='inner').reset_index().set_index('UniProt from All instances of Polypeptides in MetaCyc')
uniprots = set(uniprot_of_friendly_enz.index)
uniprot_of_friendly_enz
```

```python
from Bio import SeqIO
num_raptor_seqs = 20
records = []
with open("Dataset/uniprot-of-metacyc.faa", "r") as handle:
    for record in SeqIO.parse(handle, "fasta") :
        uniprot = record.id.split('|')[1]
        if uniprot in uniprot_of_friendly_enz.index:
            record.id = uniprot_of_friendly_enz.loc[uniprot,'index'] + '|' + record.id
            records.append( record )
for i in range(len(records) // num_raptor_seqs):
    with open("Dataset/uniprot-of-friendly-rxns-{0:03d}.faa".format(i), 'w') as out:
        SeqIO.write(records[i*num_raptor_seqs:(i+1)*num_raptor_seqs], out, "fasta")
```

```python

record_dict = SeqIO.to_dict(SeqIO.parse("Dataset/uniprot-of-metacyc.faa", "fasta"))
record_dict.keys()
```

```python
uniprot_and_pdb = pd.read_table('../MetaCyc/uniprots-and-pdb-of-MetaCyc.tab',index_col='Protein')
uniprot_and_pdb
```

```python
len(set(uniprot_and_pdb.PDB.dropna().values)), len(set(uniprot_and_pdb.PDB.dropna().index))
```

```python
import numpy as np
pdbs = set()
proteins = set()
for protein in uniprot_and_pdb.PDB.dropna().index:
    pdb_row = uniprot_and_pdb.loc[protein, 'PDB']
    proteins |= set([protein])
    pdbs |= set(pdb_row.split(','))
print('Num unique PDB entries: {}\nNum MetaCyc proteins with PDB entries: {}\nNum MetaCyc proteins that have a PDB without Uniprot: {}\n'.format(len(pdbs), len(proteins), len(proteins - set(uniprot_and_pdb.Uniprot.dropna().index))))
```
